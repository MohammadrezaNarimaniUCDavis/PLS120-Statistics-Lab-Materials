## ðŸ”„ Wrapping up on hypotheses testing:

For testing this hypothesis, we will use:
1) **Critical Value Method:** Determine the critical value(s) for a given Î± and compare with the test statistic.
2) **P-Value Method:** Calculate the p-value of the test statistic and compare with Î±.
3) **t.test function:** This method basically simplifies all the process by automatically calculating the test statistic, p-value and critical values.

The p-value represents the probability of observing a test statistic at least as extreme as the one calculated, under the null hypothesis (H0). In a two-tailed test, we are testing whether the sample mean is significantly different from the hypothesized mean (mu_0) in either direction (greater or smaller). Therefore, we compute the probability in both tails of the t-distribution and sum them. Since the t-distribution is symmetric, we only need to compute the probability of the test statistic in one tail and then double it to account for both tails. Using |t_stat| ensures we are always working with the positive magnitude, regardless of whether the test statistic is negative or positive.

## ðŸ”¬ One-sample hypotheses test: We want to compare the mean against a value.

Let's load in the data and calculate some parameters:

```r
data <- iris

# Extract the sepal length column and store it as a vector named sepal_length
sepal_length <- iris$Sepal.Length

# Define the null hypothesis mean and significance level
mu_0 <- 4 #null hypothesis mean
alpha <- 0.05 #for 95% confidence level

# Calculate sample statistics
n <- length(sepal_length)  # Sample size
sample_mean <- mean(sepal_length)  # Sample mean
sample_sd <- sd(sepal_length)  # Sample standard deviation

# Calculate the test statistic as the difference of the means divided by standard error. Remember that we are calculating the t statistic using sample stddev because the population standard deviation is not known.
t_stat <- (sample_mean - mu_0) / (sample_sd / sqrt(n))
t_stat
#this t_stat measures how far our sample_mean is from the hypothesized mean (mu_0 which is 4), in terms of the standard error of the mean.
```

**Hypothesis 1:** The mean of sepal length is NOT 4 (Two-tailed test).
- H0: Âµ = 4
- H1: Âµ â‰  4

Let's use both methods to test our hypothesis:

```r
#### 1) Critical Value Method

# Calculate the critical value for a two-tailed test with a desirable alpha level. Remember that we are calculating the t score because the population standard deviation is not known. We divide alpha by 2 because it's a two-tailed test.
t_score <- qt(1 - alpha / 2, df = n - 1)
t_score

#Make decision about the null hypothesis
abs(t_stat) > t_score 
#since the above statement is TRUE, it means that the absolute value of t statistic is greater than t score, so we can reject H0. This means that the mean of sepal length is significantly different from 4 cm.
```

```r
#### 2) p-Value Method

#calculate the p-value for a two tailed test
#The pt() function in R computes the cumulative probability up to a given t-value in the t-distribution. So if we write: pt(-abs(t_stat), df = n - 1), this gives us the area under the curve to the left of âˆ’|t_stat|. In a two-tailed test, the total p-value is the combined area in the left tail and the right tail. Since the t-distribution is symmetric, area in the right tale = area in the left tail. So the p value would be: 
p_value <- 2 * pt(-abs(t_stat), df = n - 1)

p_value 
#This p value represents the probability of obtaining a test statistic as extreme as or more extreme than t_stat, under the assumption that H0 is true. There is 8.764592e-60 probability or (8.764592e-58)% chance of observing a tstatistic => t_stat or tstatistic <= t_stat under H0. If the population mean were truly 4 (H0 = True), the chance of observing a sample mean = sample_mean (or more) is nearly impossible.


# Make decision about the null hypothesis
p_value <= alpha
#since the above statement is TRUE, we can reject H0. This means that the mean of sepal length is significantly different from 4 cm.
```

```r
#### 3) t.test function

# Perform a one-sample t-test with the alternative hypothesis: mean â‰  4
t_test_result <- t.test(sepal_length, mu = 4, alternative = "two.sided")
print(t_test_result)

#Notice how the p-value in t.test() function is different than the p-value we just calculated. The reason is that R uses a threshold for very small p-values and reports them as < 2.2e-16 instead of showing the exact value when the p-value falls below this threshold. However, both values indicate that the p-value is very small.
```

**Hypothesis 2:** The mean of sepal length is GREATER THAN 4 (One-tailed test).
- H0: Âµ â‰¤ 4
- H1: Âµ > 4

```r
#### 1) Critical Value Method

# Calculate the critical value for a one-tailed test (notice how we did not divide alpha by 2)
t_critical <- qt(1 - alpha, df = n - 1)
t_critical

# Make a decision about the null hypothesis
t_stat > t_critical
# If the above statement is TRUE, we reject H0. This means the mean of sepal length is significantly greater than 4 cm.
```

```r
#### 2) p-Value Method

# Calculate the p-value for a one-tailed test
p_value <- 1 - pt(t_stat, df = n - 1)
p_value
# This p-value represents the probability of obtaining a test statistic as extreme as t_stat or more, under the assumption that H0 is true.

# Make a decision about the null hypothesis
p_value <= alpha
# If the above statement is TRUE, we reject H0. This means that the mean of sepal length is significantly greater than 4 cm.
```

```r
#### 3) t.test function

# Perform a one-sample t-test with the alternative hypothesis: mean > 4
t_test_result <- t.test(sepal_length, mu = 4, alternative = "greater")
print(t_test_result)
```

**Hypothesis 3:** The mean of sepal length is LESS THAN 4 (One-tailed test).
- H0: Âµ â‰¥ 4
- H1: Âµ < 4

```r
#### 1) Critical Value Method

# Calculate the critical value for a one-tailed test (lower tail)
t_critical <- qt(alpha, df = n - 1)
t_critical

# Make a decision about the null hypothesis
t_stat < t_critical
# If the above statement is TRUE, we reject H0. This means the mean of sepal length is significantly less than 4 cm.
```

```r
#### 2) p-Value Method

# Calculate the p-value for a one-tailed test (lower tail)
p_value <- pt(t_stat, df = n - 1)
p_value
# This p-value represents the probability of obtaining a test statistic as extreme as t_stat or less, under the assumption that H0 is true.

# Make a decision about the null hypothesis
p_value <= alpha
# If the above statement is TRUE, we reject H0. This means that the mean of sepal length is significantly less than 4 cm.
```

```r
#### 3) t.test function

# Perform a one-sample t-test with the alternative hypothesis: mean < 4
t_test_result <- t.test(sepal_length, mu = 4, alternative = "less")
print(t_test_result)
```

## ðŸ“Š Two-sample hypotheses test (independent): We want to compare 2 mean values.

Let's calculate some parameters for each group:

```r
setosa_df <- subset(data, Species == "setosa")
versicolor_df <- subset(data, Species == "versicolor")

setosa_sepal_length <- setosa_df$Sepal.Length
versicolor_sepal_length <- versicolor_df$Sepal.Length

alpha <- 0.05

# Calculate sample statistics
n_setosa <- length(setosa_sepal_length)  # Sample size for Setosa
n_versicolor <- length(versicolor_sepal_length)  # Sample size for Versicolor
mean_setosa <- mean(setosa_sepal_length)  # Sample mean for Setosa
mean_versicolor <- mean(versicolor_sepal_length)  # Sample mean for Versicolor
sd_setosa <- sd(setosa_sepal_length)  # Sample standard deviation for Setosa
sd_versicolor <- sd(versicolor_sepal_length)  # Sample standard deviation for Versicolor
DoF <- n_setosa + n_versicolor - 2

# Calculate pooled variance for two-sample t-test (because population variance is not known. Let's imagine variances are equal)
pooled_var <- ((n_setosa - 1)*sd_setosa^2 + (n_versicolor - 1)*sd_versicolor^2) / DoF

# Calculate the test statistic
t_stat <- (mean_setosa - mean_versicolor) / sqrt(pooled_var/n_setosa + pooled_var/n_versicolor)
t_stat
```

**Hypothesis 1:** The mean of sepal length in setosa is NOT equal to sepal length in versicolor.
- H0: Âµ_setosa = Âµ_versicolor
- H1: Âµ_setosa â‰  Âµ_versicolor

```r
#### 1) Critical Value Method

# Calculate the critical value for a two-tailed test
t_critical <- qt(1 - alpha / 2, DoF)
t_critical

# Make a decision about the null hypothesis
abs(t_stat) > t_critical
# If TRUE, reject H0. This means the mean sepal lengths are significantly different between Setosa and Versicolor.
```

```r
#### 2) p-Value Method

# Calculate the p-value for a two-tailed test
p_value <- 2 * pt(-abs(t_stat), DoF)
p_value

# Make a decision about the null hypothesis
p_value <= alpha
# If TRUE, reject H0. This means the mean sepal lengths are significantly different between Setosa and Versicolor.
```

```r
#### 3) t.test Function

# Perform a two-sample t-test with unequal variances (Welch's t-test)
t_test_result <- t.test(setosa_sepal_length, versicolor_sepal_length, alternative = "two.sided")
print(t_test_result)

# Notice how the p-value in t.test() aligns with the manual p-value calculation above.
```

**Hypothesis 2:** The mean of sepal length in setosa is GREATER THAN sepal length in versicolor.
- H0: Âµ_setosa â‰¤ Âµ_versicolor
- H1: Âµ_setosa > Âµ_versicolor

```r
#### 1) Critical Value Method

# Calculate the critical value for a one-tailed test
t_critical <- qt(1 - alpha, DoF)
t_critical

# Make a decision about the null hypothesis
t_stat > t_critical
# If TRUE, reject H0. This means the mean sepal length of Setosa is significantly greater than that of Versicolor.
```

```r
#### 2) p-Value Method

# Calculate the p-value for a one-tailed test
p_value <- 1 - pt(t_stat, DoF)
p_value

# Make a decision about the null hypothesis
p_value <= alpha
# If TRUE, reject H0. This means the mean sepal length of Setosa is significantly greater than that of Versicolor.
```

```r
#### 3) t.test Function

# Perform a two-sample t-test with unequal variances (Welch's t-test)
t_test_result <- t.test(setosa_sepal_length, versicolor_sepal_length, alternative = "greater")
print(t_test_result)

# Notice how the p-value in t.test() aligns with the manual p-value calculation above.
```

**Hypothesis 3:** The mean of sepal length in setosa is LESS THAN sepal length in versicolor.
- H0: Âµ_setosa â‰¥ Âµ_versicolor
- H1: Âµ_setosa < Âµ_versicolor

```r
#### 1) Critical Value Method

# Calculate the critical value for a one-tailed test (lower tail)
t_critical <- qt(alpha, DoF)
t_critical

# Make a decision about the null hypothesis
t_stat < t_critical
# If TRUE, reject H0. This means the mean sepal length of Setosa is significantly less than that of Versicolor.
```

```r
#### 2) p-Value Method

# Calculate the p-value for a one-tailed test (lower tail)
p_value <- pt(t_stat, DoF)
p_value

# Make a decision about the null hypothesis
p_value <= alpha
# If TRUE, reject H0. This means the mean sepal length of Setosa is significantly less than that of Versicolor.
```

```r
#### 3) t.test Function

# Perform a two-sample t-test with unequal variances (Welch's t-test)
t_test_result <- t.test(setosa_sepal_length, versicolor_sepal_length, alternative = "less")
print(t_test_result)

# Notice how the p-value in t.test() aligns with the manual p-value calculation above.
```

## ðŸ” Chi-square:

1) **Test of goodness of fit:** The test of best fit tests whether there is a significant difference between an observed and expected frequency.
2) **Test of independence:** The test of best fit tests whether there is a significant difference between an observed and expected frequency.

Let's do a chi-square test for goodness-of-fit test. We want to test whether a dice is fair. If the dice is fair, we expect each of its six sides to appear equally often. We roll the dice 60 times and observe the following frequencies: (8, 12, 10, 14, 9, 7)

**Hypothesis:** The dice is not fair.
- **Null Hypothesis (H0):** The dice is fair, and the observed frequencies match the expected frequencies.
- **Alternative Hypothesis (H1):** The dice is not fair, and the observed frequencies differ from the expected frequencies.

```r
# Observed frequencies
observed <- c(8, 12, 10, 14, 9, 7)

#expected frequency for each side is 60/6 sides = 10. So we expect each side to appear 10 times.
expected <- c(10, 10, 10, 10, 10, 10)

# Perform the chi-square goodness-of-fit test
chi_square_test <- chisq.test(x = observed, p = expected / sum(expected)) #p specifies the expected proportions (10/60 for each side).

#print result
print(chi_square_test)
```

degrees of freedom is k-1 = 6-1 = 5. The p-value is not significant, so we fail to reject the null hypothesis (there is no significant evidence that the dice is unfair).

Let's do a chi-square test for independence. We want to test whether the preference for three types of fruits (Apple, Banana, Cherry) is independent of gender (Male, Female).

**Hypothesis:** fruit preference actually depends on the gender.
- **H0:** Fruit preference is independent of gender.
- **H1:** Fruit preference depends on gender.

```r
# Create the observed data as a matrix
observed_male <- c(10, 20, 30)  # Male preferences for Apple, Banana, Cherry
observed_female <- c(20, 25, 25)  # Female preferences for Apple, Banana, Cherry

observed <- matrix(c(observed_male, observed_female), nrow = 2, byrow = TRUE) #byrow = TRUE means the data fills into the matrix row by row

# Add row and column names
rownames(observed) <- c("Male", "Female")
colnames(observed) <- c("Apple", "Banana", "Cherry")

# Display the observed data
print(observed)
```

```r
# Perform the chi-square test
chi_square_test <- chisq.test(observed)

# Print the test results
print(chi_square_test)
```

The p-value is not significant, so we fail to reject the null hypothesis. There is insufficient evidence to conclude that fruit preference depends on gender.
