## ðŸ“Š Sample Size Calculation Function

Let's create a practical function for calculating sample size, building on concepts from labs 5 and 6.

**Formula:** $n = \frac{z^2 \cdot p \cdot (1-p)}{E^2}$

Where:
- $z$ = critical value from normal distribution
- $p$ = expected prevalence
- $E$ = desired margin of error

````r
calculate_sample_size <- function(prev, alpha, margin_error) {
  
  z <- qnorm(1 - alpha / 2) #or you can use qt function for t score if you need to.
  
  n <- z^2 * prev * (1 - prev) / (margin_error^2) #this was the formula for sample size
  
  required_sample_size <- ceiling(n) #this function rounds up the calculated sample size to the nearest whole number
  
  return(required_sample_size) #this will return the calculated sample size to us.
}
````

````r
## Example usage 1: step 1) Define the parameters outside of function by giving their names; step 2) use the name you assigned to the parameters inside the function
prevalence <- 0.1
significance_level <- 0.05
desired_margin_of_error <- 0.05
sample_size_example1 <- calculate_sample_size(prevalence, significance_level, desired_margin_of_error)
print(sample_size_example1)
````

````r
## Example usage 2: Use the parameter values directly inside the function
sample_size_example2 <- calculate_sample_size(prev = 0.1, alpha = 0.05, margin_error = 0.05)
print(sample_size_example2)
````

````r
## Example usage 3: Use the parameter values directly inside the function. R will read the values in order of the arguments we defined in the function.
sample_size_example3 <- calculate_sample_size(0.1, 0.05, 0.05)
print(sample_size_example3)
````

## ðŸŽ¯ Lab Objective

In this lab, we will learn to formulate hypotheses and apply two-sample t-tests to compare data between groups. Using the iris dataset, we'll compare sepal lengths between different species to determine if there are statistically significant differences between groups.

## ðŸ”¬ Hypothesis Formation

Hypothesis formulation is a critical step in statistical analysis where we define what we're testing. For the iris dataset, we'll compare mean sepal lengths between different species pairs.

For each comparison, we set up:
- **Hâ‚€:** $\mu_1 = \mu_2$ (population means are equal)
- **Hâ‚:** $\mu_1 \neq \mu_2$ (population means are different)

**Species Comparisons:**
1. Setosa vs Versicolor
2. Versicolor vs Virginica
3. Setosa vs Virginica

````r
#load the library
library(ggplot2)
library(dplyr)

#load the dataset
data <- iris

#check the summary statistic for sepal length in iris dataset
summary(data$Sepal.Length)
````

````r
#visualize the sepal length with boxplot
ggplot(data, aes(x= Species, y = Sepal.Length, fill = Species))+ geom_boxplot()
````

````r
#we'll use filter to make three data sets, each containing only one of the species. This will simplify the process of making comparisons between them later on. 
setosa_df <- data %>% filter(Species == "setosa")
versicolor_df <- data %>% filter(Species == "versicolor")
virginica_df <- data %>% filter(Species == "virginica")
````

## ðŸ“Š Two-Sample T-Test

We'll conduct two-sample t-tests assuming normal distribution and unequal variances. This uses Welch's t-test approach, which doesn't require equal population variances.

**Welch's t-test formula:** $t = \frac{\bar{x_1} - \bar{x_2}}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}$

**Key t.test() parameters:**
- `alternative = "two.sided"`: Tests for any difference between means
- `var.equal = FALSE`: Uses Welch's t-test (assumes unequal variances)
- `conf.level`: Confidence level (default 0.95)

````r
# Perform a two-sample t-test between Setosa and Versicolor
t_test_setosa_versicolor <- t.test(setosa_df$Sepal.Length, versicolor_df$Sepal.Length,
                                   alternative = "two.sided", var.equal = FALSE)

# Print the results of the t-test
print(t_test_setosa_versicolor)
````

````r
#you can also change the confidence level by setting the conf.level inside the t.test() function
t_test_setosa_versicolor <- t.test(setosa_df$Sepal.Length, versicolor_df$Sepal.Length, 
                                   alternative = "two.sided", var.equal = FALSE, conf.level = 0.99)
print(t_test_setosa_versicolor)
````

## ðŸ“ˆ Interpreting T-Test Results

Understanding t-test output is crucial for drawing valid conclusions:

- **t-statistic**: Shows direction and magnitude of difference relative to variability. Negative values indicate the first group has a lower mean.
- **p-value**: If p < Î± (typically 0.05), we reject Hâ‚€ and conclude there's a significant difference.
- **Confidence Interval**: Range of plausible values for the true mean difference. If it doesn't include 0, means are significantly different.
- **Sample Estimates**: The actual sample means for each group.

**Decision rule:** Reject Hâ‚€ if p-value < 0.05 OR if confidence interval excludes 0.

## âž¡ï¸ One-Sided Tests

Sometimes we have a specific directional hypothesis. One-sided tests are more powerful when we predict the direction of difference.

**Greater than test:** Hâ‚: Î¼â‚ > Î¼â‚‚

````r
# Perform a two-sample t-test between Setosa and Versicolor
t_test_setosa_versicolor <- t.test(setosa_df$Sepal.Length, versicolor_df$Sepal.Length,
                                   alternative = "greater", var.equal = FALSE)

# Print the results of the t-test
print(t_test_setosa_versicolor)
````

**Less than test:** Hâ‚: Î¼â‚ < Î¼â‚‚

Note: One-sided tests have more statistical power but should only be used when you have a strong theoretical reason to predict the direction.

````r
# Perform a two-sample t-test between Setosa and Versicolor
t_test_setosa_versicolor <- t.test(setosa_df$Sepal.Length, versicolor_df$Sepal.Length,
                                   alternative = "less", var.equal = FALSE)

# Print the results of the t-test
print(t_test_setosa_versicolor)
````
