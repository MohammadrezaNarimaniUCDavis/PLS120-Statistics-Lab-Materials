## Part 2: ANOVA Table Calculations (7 points)

Calculate the individual parts of the ANOVA table, then fill in the markdown table below the chunk. (5 Points)

```r
# Calculate the overall mean of weight, then add a column to the dataframe containing the overall mean
overall_mean <- mean(pigs$Weight)
cat("Overall mean weight:", overall_mean, "\n")
pigs$mean_all <- overall_mean
```

```r
# Next calculate the means of the individual treatments
diet1_df <- pigs %>% filter(Diet == "Diet.1")
diet2_df <- pigs %>% filter(Diet == "Diet.2")
diet3_df <- pigs %>% filter(Diet == "Diet.3")
diet4_df <- pigs %>% filter(Diet == "Diet.4")

diet1_mean <- mean(diet1_df$Weight)
diet1_count <- nrow(diet1_df)

diet2_mean <- mean(diet2_df$Weight)
diet2_count <- nrow(diet2_df)

diet3_mean <- mean(diet3_df$Weight)
diet3_count <- nrow(diet3_df)

diet4_mean <- mean(diet4_df$Weight)
diet4_count <- nrow(diet4_df)

# Create summary table
mean_table <- data.frame(
  Diet = c("Diet1", "Diet2", "Diet3", "Diet4"),
  Mean_Weight = c(diet1_mean, diet2_mean, diet3_mean, diet4_mean),
  Count = c(diet1_count, diet2_count, diet3_count, diet4_count)
)
print(mean_table)
```

```r
# Calculate the Sum of Squares between Groups (SSB)
SSB <- diet1_count * (diet1_mean - overall_mean)^2 +
       diet2_count * (diet2_mean - overall_mean)^2 +
       diet3_count * (diet3_mean - overall_mean)^2 +
       diet4_count * (diet4_mean - overall_mean)^2

cat("Sum of Squares Between (SSB):", SSB, "\n")
```

```r
# Calculate the Sum of Squares of the total
SST <- sum((pigs$Weight - overall_mean)^2)
cat("Total Sum of Squares (SST):", SST, "\n")
```

```r
# Calculate the Sum of Squares of the error
SSE <- SST - SSB
cat("Sum of Squares Error (SSE):", SSE, "\n")
```

```r
# Calculate the Degrees of Freedom for the treatment, the error and the total
k <- 4  # number of treatments
N <- nrow(pigs)  # total observations
df1 <- k - 1  # between groups df
df2 <- N - k  # within groups df
df3 <- N - 1  # total df

cat("Degrees of freedom between groups (df1):", df1, "\n")
cat("Degrees of freedom within groups (df2):", df2, "\n")
cat("Total degrees of freedom (df3):", df3, "\n")
```

```r
# Calculate the Mean squares between the treatments
MSB <- SSB / df1
cat("Mean Square Between (MSB):", MSB, "\n")
```

```r
# Calculate the Mean squares of the error
MSE <- SSE / df2
cat("Mean Square Error (MSE):", MSE, "\n")
```

```r
# Calculate the F statistic
F_value <- MSB / MSE
cat("F-statistic:", F_value, "\n")

# Create complete ANOVA summary table
anova_summary <- data.frame(
  Source_of_Variation = c("Between Groups", "Error", "Total"),
  Sum_of_Squares = c(SSB, SSE, SST),
  Degrees_of_Freedom = c(df1, df2, df3),
  Mean_Squares = c(MSB, MSE, NA),
  F_Value = c(F_value, NA, NA)
)
print(anova_summary)
```

Let's now confirm our result by building a linear regression model and using the anova() function, then report the p-value (2 Points)

```r
# Building a linear model between weight and diet
model <- lm(Weight ~ Diet, pigs)

# Use anova to build an anova table in R. What is the p-value? Does it match your above table?
anova_result <- anova(model)
print(anova_result)

cat("\nP-value from R's anova function:", anova_result$`Pr(>F)`[1], "\n")
cat("Does this match your manual calculations? Compare the F-values and other statistics.\n")
```
