## Question A: Calculate the mean and the 95% confidence interval for each method (4 points)

‚ùì **Calculate descriptive statistics and confidence intervals for both methods.**

```r
# Calculate mean for Method A
mean_A <- mean(tree_data$Method_A)
cat('Mean for Method A:', mean_A, 'years\n')

# Manual calculation: (22+30+26+23+23+22+23+24+22+29) / 10 = 244 / 10 = 24.4
manual_mean_A <- sum(tree_data$Method_A) / length(tree_data$Method_A)
cat('Manual calculation:', manual_mean_A, 'years')
```

```r
# Calculate mean for Method B
mean_B <- mean(tree_data$Method_B)
cat('Mean for Method B:', mean_B, 'years\n')

# Manual calculation: (20+28+22+23+21+20+20+23+18+25) / 10 = 220 / 10 = 22
manual_mean_B <- sum(tree_data$Method_B) / length(tree_data$Method_B)
cat('Manual calculation:', manual_mean_B, 'years')
```

```r
# Calculate 95% confidence interval for Method A
t_test_A <- t.test(tree_data$Method_A)
cat('Method A - Mean:', t_test_A$estimate, 'years\n')
cat('Method A - 95% CI: (', t_test_A$conf.int[1], ',', t_test_A$conf.int[2], ')\n\n')

# Manual calculation
sd_A <- sd(tree_data$Method_A)
se_A <- sd_A / sqrt(length(tree_data$Method_A))
t_critical <- qt(0.975, df = 9)  # df = n-1 = 9
ci_lower_A <- mean_A - t_critical * se_A
ci_upper_A <- mean_A + t_critical * se_A

cat('Manual calculation:\n')
cat('Standard deviation:', round(sd_A, 2), '\n')
cat('Standard error:', round(se_A, 2), '\n')
cat('t-critical (df=9, Œ±=0.05):', round(t_critical, 3), '\n')
cat('95% CI: (', round(ci_lower_A, 3), ',', round(ci_upper_A, 3), ')')
```

```r
# Calculate 95% confidence interval for Method B
t_test_B <- t.test(tree_data$Method_B)
cat('Method B - Mean:', t_test_B$estimate, 'years\n')
cat('Method B - 95% CI: (', t_test_B$conf.int[1], ',', t_test_B$conf.int[2], ')\n\n')

# Manual calculation
sd_B <- sd(tree_data$Method_B)
se_B <- sd_B / sqrt(length(tree_data$Method_B))
ci_lower_B <- mean_B - t_critical * se_B
ci_upper_B <- mean_B + t_critical * se_B

cat('Manual calculation:\n')
cat('Standard deviation:', round(sd_B, 2), '\n')
cat('Standard error:', round(se_B, 2), '\n')
cat('95% CI: (', round(ci_lower_B, 3), ',', round(ci_upper_B, 3), ')')
```

## Question C: Calculate the difference between the means of the two methods and determine its 95% confidence interval (4 points)

‚ùì **Calculate the mean difference and its confidence interval.**

```r
# Calculate the difference between means
diff_means <- mean_A - mean_B
cat('Difference in means (Method A - Method B):', diff_means, 'years\n')

# Manual calculation: 24.4 - 22 = 2.4
cat('Manual calculation: 24.4 - 22 =', diff_means, 'years')
```

```r
# Calculate 95% confidence interval for the difference (independent samples)
t_test_diff <- t.test(tree_data$Method_A, tree_data$Method_B, paired = FALSE)
cat('Difference in means:', t_test_diff$estimate[1] - t_test_diff$estimate[2], 'years\n')
cat('95% CI for difference: (', t_test_diff$conf.int[1], ',', t_test_diff$conf.int[2], ')\n\n')

# Manual calculation
pooled_var <- ((9 * sd_A^2) + (9 * sd_B^2)) / 18
se_diff <- sqrt((sd_A^2)/10 + (sd_B^2)/10)
t_critical_18 <- qt(0.975, df = 18)  # df = n1 + n2 - 2 = 18
ci_lower_diff <- diff_means - t_critical_18 * se_diff
ci_upper_diff <- diff_means + t_critical_18 * se_diff

cat('Manual calculation:\n')
cat('Pooled variance:', round(pooled_var, 3), '\n')
cat('SE of differences:', round(se_diff, 3), '\n')
cat('t-critical (df=18):', round(t_critical_18, 4), '\n')
cat('95% CI: (', round(ci_lower_diff, 3), ',', round(ci_upper_diff, 3), ')')
```

## Question E: Perform a two-sample t-test to evaluate whether the two methods differ (4 points)

‚ùì **Form your hypotheses and perform an independent samples t-test.**

```r
# Perform two-sample t-test (independent samples)
t_test_independent <- t.test(tree_data$Method_A, tree_data$Method_B, paired = FALSE)
print(t_test_independent)

# Manual calculation
t_statistic <- diff_means / se_diff
cat('\nManual calculation:\n')
cat('t-statistic:', round(t_statistic, 3), '\n')
cat('Critical t-value (df=18, Œ±=0.05):', round(t_critical_18, 4), '\n')

if (abs(t_statistic) > t_critical_18) {
  cat('Decision: Reject H‚ÇÄ (significant difference)\n')
} else {
  cat('Decision: Fail to reject H‚ÇÄ (no significant difference)\n')
}

cat('Since |t| =', round(abs(t_statistic), 3), 'is not greater than t-critical =', 
    round(t_critical_18, 4), ', we fail to reject the null hypothesis.')
```

## Question F: Perform a paired t-test to evaluate whether the two methods differ (4 points)

‚ùì **Form your hypotheses and perform a paired samples t-test.**
